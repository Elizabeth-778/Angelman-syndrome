```python
outliers_indices_dict = {}
for gene in normalized_df.columns:

    Q1 = normalized_df[gene].quantile(0.25)
    Q3 = normalized_df[gene].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_mask = (normalized_df[gene] < lower_bound) | (normalized_df[gene] > upper_bound)

    outlier_indices = normalized_df[gene][outliers_mask].index.tolist()

    if outlier_indices:
        outliers_indices_dict[gene] = outlier_indices
```

```python
from collections import Counter

all_indices = [index for indices in outliers_indices_dict.values() for index in indices]
index_frequency = Counter(all_indices)
most_common_indices = index_frequency.most_common()

print("Most common indices and their frequencies:")
for index, count in most_common_indices:
    print(f"Index: {index}, Frequency: {count}")
```

```python
accuracy_scores = []
y = normalized_df['Disease status']
X = normalized_df.drop(columns=['Disease status'], axis=1).values
model = LogisticRegression(max_iter=1000)

for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

average_accuracy = np.mean(accuracy_scores)
print("Average Jackknife Accuracy:", average_accuracy)
```
